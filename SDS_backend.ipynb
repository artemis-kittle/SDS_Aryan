```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("EdX.csv")
data.head(5)
```

```python
data = data[['Name','University','Difficulty Level','Course Description']]
```

```python
data.head(5)
```

```python
data['Name'] = data['Name'].str.replace(' ',',')
data['Name'] = data['Name'].str.replace(',,',',')
data['Name'] = data['Name'].str.replace(':','')
data['Course Description'] = data['Course Description'].str.replace(' ',',')
data['Course Description'] = data['Course Description'].str.replace(',,',',')
data['Course Description'] = data['Course Description'].str.replace('_','')
data['Course Description'] = data['Course Description'].str.replace(':','')
data['Course Description'] = data['Course Description'].str.replace('(','')
data['Course Description'] = data['Course Description'].str.replace(')','')
```

```python
data.head(5)
```

```python
data['tags'] = data['Name'] + data['University'] + data['Difficulty Level'] + data['Course Description']
```

```python
data.head(5)
```

```python
data['tags'].iloc[1]
```

```python
new_df = data[['Name','tags']]
new_df.head(5)
```

```python
new_df['tags'] = data['tags'].str.replace(',',' ')
new_df['Name'] = data['Name'].str.replace(',',' ')
new_df.rename(columns = {'Name':'course_name'}, inplace = True)
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower()) 
```

```python
new_df.head(5)
```

```python
new_df.shape
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
    
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem) 
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)
```

```python
def result(course):
    course_index = new_df[new_df['course_name'] == course].index[0]
    distances = similarity[course_index]
    course_list = sorted(list(enumerate(distances)),reverse=True, key=lambda x:x[1])[1:6]
    p=0
    for i in course_list:
      if(p<6):
       print(new_df.iloc[i[0]].course_name)
       p+=1
```

```python
result('The Foundations of Happiness at Work') 
```
